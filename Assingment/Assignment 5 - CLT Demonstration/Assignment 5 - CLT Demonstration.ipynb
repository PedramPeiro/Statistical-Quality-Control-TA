m should be greater than 20 or 25 (and in theory should tend to infinity), in the other hand, n is not limited but its better to be greater than 10

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import norm
```

```python
# Binomial cacl
m=500
n=10
arr_binom = np.zeros(m)
for i in range(0,m):
    arr_binom[i]=np.mean(np.random.binomial(100 , 0.3 , size=n))
```

```python
# Gamma cacl
arr_gamma = np.zeros(m)
for i in range(0,m):
    arr_gamma[i]=np.mean(np.random.gamma(2 , 5 , size=n))
```

```python
# Uniform cacl
arr_uniform = np.zeros(m)
for i in range(0,m):
    arr_uniform[i]=np.mean(np.random.uniform(5 , 10 , size=n))
```

```python
# Geom cacl
arr_geom = np.zeros(m)
for i in range(0,m):
    arr_geom[i]=np.mean(np.random.geometric(0.3 , size=n))
```

```python
# function for normal calculations

def normal_calc( in_arr):
    mu = np.mean(in_arr)
    std = np.std(in_arr)
    min_bin=min(in_arr)
    max_bin=max(in_arr)
    x = np.linspace( min_bin , max_bin , 200)
    arr_norm = norm.pdf( x , mu , std)
    
    return (x,arr_norm)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
plt.style.use('seaborn')

sns.kdeplot(ax=ax1 , x=arr_binom , color='dodgerblue' , label='binomial' , ls='--')
ax1.plot(normal_calc(arr_binom)[0] , normal_calc(arr_binom)[1] , color='tomato' , label='normal'  ,ls=':')

sns.kdeplot(ax=ax2 , x=arr_gamma , color='dodgerblue' , label='gamma' , ls='--')
ax2.plot(normal_calc(arr_gamma)[0] , normal_calc(arr_gamma)[1] , color='tomato' , label='normal'  ,ls=':')

sns.kdeplot(ax=ax3 , x=arr_uniform , color='dodgerblue' , label='uniform' , ls='--')
ax3.plot(normal_calc(arr_uniform)[0] , normal_calc(arr_uniform)[1] , color='tomato' , label='normal'  ,ls=':')

sns.kdeplot(ax=ax4 , x=arr_geom , color='dodgerblue' , label='geometric' , ls='--')
ax4.plot(normal_calc(arr_geom)[0] , normal_calc(arr_geom)[1] , color='tomato' , label='normal'  ,ls=':')

ax=[ax1 , ax2 , ax3 , ax4]
titles=['binom' , 'gamma' , 'uniform' , 'geometric']
[x.set_title(y) for x,y in zip(ax , titles)]
[x.legend() for x in ax]
plt.tight_layout()
plt.show()
```

# Q2

```python
years_lis=[str(x) for x in range(2006 , 2016)]
years_lis.append('Country')
df_GDP=df.loc[: ,years_lis ]
df_GDP['total_average']=np.mean(df_GDP.iloc[: , :10] , axis=1)
df_GDP=df_GDP.set_index('Country').sort_values('total_average',ascending=False)
df_ans2=df_GDP.drop([str(x) for x in range(2006 , 2016)],axis=1)
df_ans2
```

# Q3

```python
years_lis=[str(x) for x in range(2006 , 2016)]
plt.figure()
plt.style.use('seaborn')
plt.plot(years_lis , df_GDP.loc['United States',years_lis] , label='United States',c='dodgerblue')
plt.plot(years_lis , df_GDP.loc['China',years_lis] , label='United States',c='tomato')
plt.plot(years_lis , df_GDP.loc['Japan',years_lis] , label='United States',c='red')
plt.legend()
plt.title('GDP of superior countries')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.tight_layout()
plt.show()
```

# Q4

```python
df_energy=df.loc[: , ['Country','Energy Supply' , 'Energy Supply per Capita']]
df_energy['Population']=df_energy['Energy Supply']/df_energy['Energy Supply per Capita']
df_energy=df_energy.sort_values('Population',ascending=True).iloc[:5 ,[0 , 3] ]

fig , ax=plt.subplots()
slices=df_energy['Population'].to_list()
labels=df_energy['Country'].to_list()
ax.pie(slices , labels=labels , shadow=True , startangle=90 , autopct='%1.1f%%' , wedgeprops={'edgecolor':'black'})
ax.set_aspect('equal')
ax.set_title('Population' , fontsize=18)
fig.tight_layout()
plt.show()
```

# Q5

```python
df_cite=df.iloc[: , :8]
df_cite['Ratio']=df_cite['Self-citations']/df_cite['Citable documents']
df_cite=df_cite.sort_values('Ratio',ascending=False)
fig , ax=plt.subplots()
ax.bar(df_cite['Country'] , df_cite.Ratio )
ax.set_xticks(range(0,len(df_cite)))
ax.set_xticklabels(labels=df_cite['Country'],fontsize=8,rotation=45);
for i in range(0,len(df_cite)):
    ax.text(i-0.35 , df_cite.iloc[i ,8]+0.05,np.around(df_cite.iloc[i ,8],decimals=2))
ax.set_title('Ratio')
plt.tight_layout()
plt.show()
```
